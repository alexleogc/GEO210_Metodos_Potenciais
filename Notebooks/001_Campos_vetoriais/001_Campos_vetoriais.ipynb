$$
\nabla \times \overrightarrow{\mathrm{F}}=\left(\frac{\partial \mathrm{F}_{\mathrm{z}}}{\partial \mathrm{y}}-\frac{\partial \mathrm{F}_{\mathrm{y}}}{\partial \mathrm{z}}\right) \overrightarrow{\mathrm{i}}+\left(\frac{\partial \mathrm{F}_{\mathrm{x}}}{\partial \mathrm{z}}-\frac{\partial \mathrm{F}_{\mathrm{z}}}{\partial \mathrm{x}}\right) \overrightarrow{\mathrm{j}}+\left(\frac{\partial \mathrm{F}_{\mathrm{y}}}{\partial \mathrm{x}}-\frac{\partial \mathrm{F}_{\mathrm{x}}}{\partial \mathrm{y}}\right) \overrightarrow{\mathrm{k}}
$$

```python
def curl(F):
    c1 = diff(F[2],y) - diff(F[1],z)
    c2 = diff(F[0],z) - diff(F[2],x)
    c3 = diff(F[1],x) - diff(F[0],y)
    return [c1,c2,c3]
```

####  Plotando grficos 2D

### Plotando a funo

$$ \mathbf{F}(x,y) = \frac{x}{\sqrt{x^2+y^2}}\mathbf{i} + \frac{y}{\sqrt{x^2+y^2}}\mathbf{j} $$

```python
#criando malha e determinando a funo
x_min, x_max = -10,10
y_min, y_max = -10,10
n_ams_x, n_ams_y = 11, 11

x,y = np.meshgrid(np.linspace(x_min,x_max,n_ams_x),
                  np.linspace(y_min,y_max,n_ams_y))

fx = x/np.sqrt(x**2 + y**2)
fy = y/np.sqrt(x**2 + y**2)

# fx = x*y
# fy = x**2
```

```python
fig,axes = plt.subplots(figsize=(7,7))

axes.quiver(x,y,fx,fy)
```

### Plotando campos 3D 

### Plotando a funo

$$ \mathbf{F}(x,y,z) = \frac{x}{{x^2+y^2+z^2}}\mathbf{i} + \frac{y}{{x^2+y^2+z^2}}\mathbf{j} +\frac{z}{{x^2+y^2+z^2}}\mathbf{k} $$

```python
#criando malha e determinando a funo
x_min, x_max = -2,2
y_min, y_max = -2,2
z_min, z_max = -2,2
n_ams_x, n_ams_y, n_ams_z = 5, 5, 5

x,y,z = np.meshgrid(np.linspace(x_min,x_max,n_ams_x),
                  np.linspace(y_min,y_max,n_ams_y),
                   np.linspace(z_min,z_max,n_ams_z))

fx = x/(x**2 + y**2 + z**2)
fy = y/(x**2 + y**2 + z**2)
fz = z/(x**2 + y**2 + z**2)
```

```python
fig = plt.figure(figsize=(8,8))
axes = fig.gca(projection='3d')

axes.quiver(x,y,z,fx,fy,fz)
```

### Obtendo o campo vetorial a partir da funo gradiente

$$\mathbf{F} = \nabla \phi$$

```python
#criando malha e determinando a funo
x_min, x_max = -10,10
y_min, y_max = -10,10
n_ams_x, n_ams_y = 11, 11

x,y = np.meshgrid(np.linspace(x_min,x_max,n_ams_x),
                  np.linspace(y_min,y_max,n_ams_y))

phi = -1/np.sqrt(x**2 + y**2)

#phi = x**2

fy,fx = np.gradient(phi)
```

```python
fig,axes = plt.subplots(figsize=(7,7))

axes.quiver(x,y,fx,fy)
```
